```python
import yfinance as yf

cdp = yf.download("CDR.WA", start="2020-01-01", end="2024-12-31")
cdp.head()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12,6))
plt.plot(cdp['Close'])
plt.title("CD Projekt  Cena zamknicia (20202024)")
plt.xlabel("Data")
plt.ylabel("Cena (PLN)")
plt.grid(True)
plt.show()
```

```python
srednia = round(cdp['Close'].mean(), 2)
odchylenie = round(cdp['Close'].std(), 2)

print("rednia cena:", srednia)
print("Odchylenie standardowe:", odchylenie)
```
